```python
def print_null_percentage(data):
    total_rows = len(data)
    null_counts = data.isnull().sum()
    null_percentages = (null_counts / total_rows) * 100
    
    for column, percentage in null_percentages.iteritems():
        print(f"Column: {column}\t\tNull Percentage: {percentage:.2f}%")

# Call the function for merged_train_data
print("Null percentages for merged_train_data:")
print_null_percentage(merged_train_data)
print()

# Call the function for merged_test_data
print("Null percentages for merged_test_data:")
print_null_percentage(merged_test_data)
```
